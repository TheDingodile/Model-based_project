```python
import pandas as pd
import numpy as np
```

```python
lifedf= pd.read_csv('/home/gzn/pgm/LED_continent.csv')
```

```python
gdp=pd.read_excel('/home/gzn/pgm/API_NE.GDI.TOTL.KD_DS2_en_excel_v2_5463459.xls',sheet_name='Data')
```

```python
gdp= gdp[['Country Name','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']]
```

```python
gdp.set_index('Country Name',inplace=True)
```

```python
nan_gdp = lifedf[lifedf['GDP'].isna()]
nan_gdp
```

```python
missdf = nan_gdp[['Country','Year']]
for i in range(len(missdf)):
    country = str(missdf.iloc[i,0])
    year = str(missdf.iloc[i,1])
    if country in set(gdp.index.tolist()):
        nan_gdp.iloc[i,18]=gdp.loc[country,year]
    else:
        nan_gdp.iloc[i,18]=np.nan
```

```python
nan_gdp = nan_gdp.replace('NaN', np.nan)
```

```python
nana = nan_gdp[nan_gdp['GDP'].isna()]
len(nana)
```

```python
gdpdropna = nan_gdp.dropna(subset=['GDP'])
gdpdropna
```

```python
lifedf.update(gdpdropna)
```

```python
pop=pd.read_excel('/home/gzn/pgm/API_SP.POP.TOTL_DS2_en_excel_v2_5454877.xls',sheet_name='Data')
pop= pop[['Country Name','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']]
```

```python
pop.set_index('Country Name',inplace=True)
nan_pop = lifedf[lifedf['Population'].isna()]
nan_pop
```

```python
missdf = nan_pop[['Country','Year']]
for i in range(len(missdf)):
    country = str(missdf.iloc[i,0])
    year = str(missdf.iloc[i,1])
    if country in set(pop.index.tolist()):
        nan_pop.iloc[i,19]=pop.loc[country,year]
    else:
        nan_pop.iloc[i,19]=np.nan
```

```python
nan_pop = nan_pop.replace('NaN', np.nan)
nana = nan_pop[nan_pop['Population'].isna()]
len(nana)
```

```python
popdropna = nan_pop.dropna(subset=['Population'])
popdropna
```

```python
for i in range(len(popdropna)):
    index = int(popdropna.iloc[1]['Unnamed: 0'])
    popdropna.iloc[i,18]=lifedf.iloc[index,18]
```

```python
lifedf.update(popdropna)
```
